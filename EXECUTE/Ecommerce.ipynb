```python
# Importando dados
import random
import datetime
from faker import Faker 
import pandas as pd
```

```python
#Definindo parâmetros
fake = Faker("pt_BR")
numeros_registros = 1000000
nome_arquivo = "Ecommerce_2024.csv"
caminho_saida = f"C:/Users/felip/Documents/Code/Cursos/E-COMMERCE/PBI_Version/"
```

```python
# Criando base para os dados
colunas = [
    "Nome", "Sobrenome","Sexo", "CPF","Idade", "ID_Produto", "Produto", "Quantidade",
    "Valor_Unitario", "Total_Compra", "Imposto_Produto", "Cidade",
    "Estado", "Pais", "Data", "Forma_Pagamento"]

produtos = [{"id": "01", "nome": "Notebook", "valor": 3500.00},
            {"id": "02", "nome": "Smartphone", "valor": 2800.00},
            {"id": "03", "nome": "Tablet", "valor": 3500.80},
            {"id": "04", "nome": "Impressora", "valor": 800.00},
            {"id": "05", "nome": "Monitor", "valor": 950.80},
            {"id": "06", "nome": "Teclado", "valor": 200.90},
            {"id": "07", "nome": "Mouse", "valor": 89.90},
            {"id": "08", "nome": "HeadSet", "valor": 250.00},
            {"id": "09", "nome": "Cadeira Gamer", "valor": 980.00},
            {"id": "10", "nome": "Suporte Monitor", "valor": 150.99}]

formas_pagamento = ["Debito", "Credito", "PIX", "Especie"]

sexo_opcoes = ["Masculino", "Feminino"]

dados = []
```

```python
#Loop para preencher os dados
for i in range(0, numeros_registros):
    sexo = random.choice(sexo_opcoes)
    nome = fake.first_name_male() if sexo == "Masculino" else fake.first_name_female()
    sobrenome = fake.last_name()
    idade = random.randint(18, 60)
    cpf = fake.cpf()

    #Selecionando Produto
    produto_selecionado = random.choice(produtos)

    id_produto = produto_selecionado["id"]
    nome_produto = produto_selecionado["nome"]
    valor_unitario = produto_selecionado["valor"]

    #Quantidade de itens
    quantidade = random.randint(1, 5)
    total_compra = round(quantidade * valor_unitario, 2)

    imposto_produto = round(total_compra * 0.10, 2)

    cidade = fake.city()
    estado = fake.estado_sigla()
    pais = "Brasil"

    #Configurando range de datas
    data_inicial = datetime.datetime(2024, 1, 1)
    data_final = datetime.datetime(2024, 12, 31)

    data_compra = fake.date_between(start_date= data_inicial,end_date=data_final)
    pagamento = random.choice(formas_pagamento)

    dados.append([nome, sobrenome,sexo,cpf,idade,id_produto,nome_produto,quantidade,valor_unitario,total_compra,imposto_produto,cidade,estado,pais,data_compra,pagamento])
```

```python
#Criando DataFrame
df = pd.DataFrame(data= dados, columns=colunas)
```
